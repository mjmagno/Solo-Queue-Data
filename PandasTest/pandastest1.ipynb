```python
my_puuid = "qkM3HGY7VhJbh3lZFNSgWknUyJGfe9NnTmbsSLppeVCKyg9w0gbryOeyG1h-pNFV7n_e0EEwYsBmtA"
def getParticipantNumber(puuid):
    for i, participant_puuid in enumerate(data["metadata"]["participants"]):
        if participant_puuid == puuid:
            return i + 1  # Add 1 to match the participant number system

part_number = getParticipantNumber(my_puuid)
print(part_number)
```

```python
# Normalize the nested JSON structure (for example, participant data)
participants_df = pd.json_normalize(data['info']['frames'][0]['participantFrames'][part_number])
participants_df.head()
```

```python
# Function to extract all frames for a given participant number
def getParticipantFrames(participant_number, data):
    frames_data = []
    
    # Loop through each frame
    for frame in data['info']['frames']:
        # Extract the participant's frame data based on their number
        participant_frame = frame['participantFrames'].get(str(participant_number), {})
        # TODO frame['events]['timestamp]
        participant_frame['timestamp'] = frame['timestamp']  # Add the timestamp for context
        frames_data.append(participant_frame)
    
    # Convert the list of frame data to a DataFrame
    return pd.DataFrame(frames_data)

df = getParticipantFrames(part_number,data)
df.head()
```

```python
# Normalize the nested JSON structure (for example, participant data)
participants_df = pd.json_normalize(match_data['info']['participants'][part_number - 1])
participants_df.head()
```

```python
def findOpponent(my_puuid):
    for participant in match_data["info"]["participants"]:
        if participant["puuid"] != my_puuid and participant["individualPosition"] == "TOP":
            return pd.json_normalize(participant)

opponent_df = findOpponent(my_puuid)
opponent_df.head()
```
